## Premise of the Guessing Game
1. Python generates a random secret number between 1-20.
1. The player has 5 tries to guess the secret number (a 25% chance to guess it correctly).
1. After each player guess, Python tells the player if the player's guess is higher or lower than the secret number. 
1. The game ends when
    1. the player uses all five tries without a correct guess or
    1. the player guesses the correct number.
    
## Scoring
- If the player guesses the secret number in 1 try, they receive 5 points.
- If the player guesses the secret number in 2 tries, they receive 4 points.
- If the player guesses the secret number in 3 tries, they receive 3 points.
- If the player guesses the secret number in 4 tries, they receive 2 points.
- If the player guesses the secret number in 5 tries, they receive 1 point.
- If the player does not guess the secret number, they receive 0 points. 

The fewer tries to correctly guess the secret number, the higher the player scores.

## At the End of each Game
At the end of each game, the following is recorded to a DataFrame:
- Player's name
- Game number for that player
- Player's score
- Secret number for that game

## At the End of all of the Player's Games
When the player is done playing all the games they want to play, a chart is automatically generated showing
- That player's name
- That player's highest score
- That player's lowest score
- That player's median score
- The statistical distribution of all scores for that player to date

Then, save the DataFrame to a csv file. 

## For a bit more of a challenge

```python

# ----- initialize some needed data structures -----

# The Dataframe where we will store results
df = pd.read_csv('data/p2cgame.csv', index_col=0)

stats = {}
game_times = {}

# -------------------------
# ----- The Game Code -----

# ----- Opening Greeting -----
name = input('Well, Hello! What is your name? ')
print('\nI am excited to play "The Guessing Game". I will think of a random number between 1-20,')
print('then I will give you five tries to guess the number. To be fair, I will tell you if')
print('the number you guessed is higher or lower than the number of which I am thinking.\n ')


# ----- Determine if a player has played before and is in the files -----
if name in (list(set(list(df['Player_name'])))):
    gk = df.groupby('Player_name') #Group DF entries by the player's name
    gkk = gk.get_group(name) #Assign the group_by to a variable
    game_number = max(gkk['Game_number']+1)#Determine the number of games of the player
else:
    print(name, "hasn't played before.")
    game_number = 1
    

# ----- Initialize a non-conventional function -----
play_again = input(name +', would you like to play "The Guessing Game"? yes or no ')

while play_again == 'yes' or play_again == 'Yes':
    # Generate an initial random secret number
    secret_number = random.randint(1, 20)
    number_of_tries = 1
    score = 5
    df.loc[(game_number+1),'Game_number'] = game_number
    while number_of_tries < 6:
        user_guess = input('\nGo ahead, ' + name + ' and guess a number. \n')
        user_guess = int(user_guess)
        
        if user_guess < secret_number:
            print('Good try, ' + name + ', but you are too low.\n')
            score -= 1
            number_of_tries += 1
        elif user_guess > secret_number:
            print('Good try, ' + name + ', but you are too high.\n')
            score -= 1
            number_of_tries += 1
        elif user_guess == secret_number:
            print('PERFECT, '+ name + '! You guessed it. \n')
            game_times['game_'+str(game_number)] = number_of_tries
            score -= 1
            number_of_tries +=1
            break
        else: 
            print("I'm sorry, " + stats['name'] + ". You didn't get it.\n ")
            break
            
           
        
    # ----- Give the player the answer -----
    print('The number I was thinking of was '+ str(secret_number) + ".\n")
    
    game_number += 1 # Add 1 to the game number 
    random_number = random.randint(1, 20)# Select the secret number for the next game
    
    # ----- Would the player like to play again -----
    play_again = input(name + ', would you like to play again? ')
    
    # ----- Save the data -----
    df.loc[len(df.Player_name), 'Player_name'] = name # Record player's name
    df.loc[(len(df.Game_number)-1), 'Game_number'] = game_number # Record game number
    df.loc[(len(df.Player_score)-1), 'Player_score'] = score
    df.loc[(len(df.Secret_number)-1), 'Secret_number'] = secret_number # Record the random secret number
    df.to_csv('data/p2cgame.csv')
    

# ----- Farewell text -----
print('\nThank you, '+ name +'. It was a pleasure to meet you.' )
print('You played the game '+ str(len(game_times))+' times.')

# ----- Text if the player does not play the game initially -----
if len(game_times) == 0:
    print("You didn't play the game, so you don't have a score.")
    

# ---------- PLOTTING --------
# ----- Import necessary libraries for plotting -----            
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# ----- Making a Boxplot -----
plt.figure(figsize = (12,8))# Specify the size of the figure so it is easier to read
sns.boxplot(data = df, x = df.Player_name, y = df.Player_score)# Actual plotting command
plt.title('Score statistics')# Add a title
plt.xlabel('Player')# Add an x axis label
plt.ylabel('Score')# Add a y axis label
plt.ylim(0, 5) # Specify the y axis scale

# ----- Making a Violin plot -----
df.Player_name = df.Player_name.astype(str)# Insure Player_name is string
df.Player_score = df.Player_score.astype(float)# Insure Player_score is integer
df.Game_number = df.Game_number.astype(float)# Insure Game_number is integer
df.Secret_number = df.Secret_number.astype(float)# Insure Secret_number is integer

plt.figure(figsize = (12, 8))# Specify the size of the figure so it is easier to read

# Actual plotting Command
sns.violinplot(data = df, 
               x = 'Player_name', 
               y = 'Player_score', 
              )

plt.title("Box plots of Player Scores based on Player Name")# Add a title
plt.xlabel("Player")# Add an x axis label
plt.ylabel("Score")# Add a y axis label
plt.ylim(0, 5)# Specify the y axis scale

plt.show()# Command to show the graphs
```
